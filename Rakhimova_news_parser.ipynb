```python
import requests as req
from bs4 import BeautifulSoup
```

```python
url = "https://news.vtomske.ru"
r = req.get(url)
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html')
print(soup.prettify())
```

```python
one_page_a = soup.findAll('a', class_ ="news-small")
```

```python
len(soup.findAll('a', class_ ="news-small"))
```

```python
one_page_links = ["https://news.vtomske.ru"+i.get("href") for i in one_page_a]
```

```python
one_page_links
```

```python
one_page_list = []
for page in one_page_links:
  r = req.get(page)
  soup = BeautifulSoup(r.text, 'html')
  title = soup.find("h1", class_ = "material-title").text
  # text = soup.find("div", class_ = "full-text").text # собрать текст
  text = [i.text for i in soup.find("div", class_="full-text").findAll("p")]
  text = ' '.join(text)
  author = soup.find("div", class_ = "author").text
  datetime = soup.find("time").get("datetime")
  #tags = soup.find("div", class_="info").findAll("a", class_="tag")
  tags = [i.text for i in soup.find("div", class_="info").findAll("a", class_="tag")]
  print(page, title, text, author, datetime, tags, sep = "\n")
  one_page_list.append({"url": page, "title": title, "text": text, "author": author, "datetime": datetime, "tags": tags})
```
