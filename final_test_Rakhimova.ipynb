```python
import requests as req
from bs4 import BeautifulSoup
```

```python
url = "https://www.kinopoisk.ru/lists/movies/best_movies_2022_editorialchoice/"
r = req.get(url)
print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'html')
print(soup.prettify())
```

```python
link = 'https://www.kinopoisk.ru'+soup.find('div', class_='styles_root__ti07r').find('a', class_='styles_root__wgbNq').get('href')
link
```

```python
russian_name = soup.find('div', class_='styles_root__ti07r').find('span').text
russian_name
```

```python
english_name = soup.find('div', class_='styles_root__ti07r').find('div', class_='desktop-list-main-info_secondaryTitleSlot__mc0mI').find('span').text
english_name
```

```python
film_type = soup.find('div', class_='styles_root__ti07r').find('div', class_='desktop-list-main-info_additionalInfo__Hqzof').find('span').text
film_type
```

```python
cast = soup.find('div', class_='styles_root__ti07r').findAll('span', class_='desktop-list-main-info_truncatedText__IMQRP')[1].text
cast
```

```python
rate = soup.find('div', class_='styles_root__ti07r').find('div', class_='styles_kinopoiskValueBlock__qhRaI').find('span').text
rate
```

```python
films = soup.findAll('div',class_='styles_root__ti07r')
len(films)
```

```python
films = soup.findAll('div',class_='styles_root__ti07r')

data = []

for film in films:
  link = 'https://www.kinopoisk.ru'+film.find('a', class_='styles_root__wgbNq').get('href')
  russian_name = film.find('span').text
  english_name = film.find('div', class_='desktop-list-main-info_secondaryTitleSlot__mc0mI').find('span').text
  film_type = film.find('div', class_='desktop-list-main-info_additionalInfo__Hqzof').find('span').text
  cast = film.findAll('span', class_='desktop-list-main-info_truncatedText__IMQRP')[1].text
  rate = film.find('div', class_='styles_kinopoiskValueBlock__qhRaI').find('span').text



  data.append([link, russian_name, english_name, film_type, cast, rate])
```

```python
data
```

```python
len(data)
```

```python
import pandas as pd
```

```python
header = ['link', 'russian_name', 'english_name', 'film_type', 'cast', 'rate']
```

```python
df = pd.DataFrame(data, columns=header)
```
